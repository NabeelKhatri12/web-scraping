```python
import requests
import urllib.request
import shutil
from bs4 import BeautifulSoup
import os
import xlsxwriter
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
```

```python
link = []
for page in range(1, 8):
    url = 'https://www.argos.co.uk/browse/appliances/laundry/tumble-dryers/c:29614/opt/page:{}'.format(page)
    response = requests.get(url, headers = headers)
    response
    html_soup = BeautifulSoup(response.text, 'html.parser')
    aas = html_soup.find_all("a", class_='SpecCardstyles__Title-sc-2rwnrs-5 fXJBMz')
    for root in aas:
        link.append(root.get("href"))
```

```python
url = []
for i in range(0, len(link)):
    hUrl = "https://www.argos.co.uk" + link[i]
    url.append(hUrl)
```

```python
image_info = []
src = []
prod_title = []
pro_title = []
price_list = []
pr_list = []
desc_list = []
desc_li = []
specs_list = []
specs_li = []
workbook = xlsxwriter.Workbook('C:/Users/Dua/Desktop/tumble_dryers1Data.xlsx')
worksheet = workbook.add_worksheet()
cell_format = workbook.add_format()
cell_format.set_text_wrap()
for re in range(30, 60):
    response = requests.get(url[re], timeout = 5, headers = headers)
    html_soup = BeautifulSoup(response.text, 'html.parser')
    image = html_soup.find_all("div", class_='MediaGallerystyles__ImageWrapper-sc-1jwueuh-2 bhjltf')
    product_title = html_soup.find_all("div", class_='Namestyles__ProductName-sc-269llv-0 fZbWJT bolt-v2')
    price = html_soup.find_all("li", class_='Pricestyles__Li-sc-1oev7i-0 haYQtv')
    desc = html_soup.find_all("div", class_='Descriptionstyles__Container-sc-1xg1hca-0 gmCmHe')
    specs = html_soup.find_all("div", class_ = 'SpecificationTablestyles__SpecificationTable-sc-171ywi2-0 hdOeaG')

#for image

    for a in image:
        image_tag = a.findChildren("img")
        image_info.append((image_tag[0]["src"]))
    for links in range(0, len(image_info)):
        #image_url = image_info[links]
        i_url = "https:" + image_info[links]
    
    for j in range(0,2):
        src.append(i_url)
        
    def download_image(image):
        response = requests.get(image, stream=True, headers = headers, timeout = 5)
        #realname = ''.join(e for e in image[0:] if e.isalnum())
        #name = str(re)
        #for k in range(97, 111):
         #   img_name = chr(k) + name
          #  if name[-1] = "9":
           #     k++
#Path change krna iss ka
        file = open("C:/Users/Dua/Desktop/argos/tumble_dryers1/{}.jpg".format(re), 'wb')

        response.raw.decode_content = True
        shutil.copyfileobj(response.raw, file)
        del response
        
    for i in range(0, len(src)):
        download_image(src[i])

#Excel code
    #resizing cells
    worksheet.set_column('A1:A5', 80)
    worksheet.set_default_row(250)

    #Path of the direcory where images are saved
    path = 'tumble_dryers1'
    root_path = 'C:/Users/Dua/Desktop/argos/tumble_dryers1'
    files = os.listdir(path)
    fileSorted = sorted(files,key=lambda x: int(os.path.splitext(x)[0]))
    img_path = []
    for f in fileSorted:
        img_path.append((os.path.join(root_path, f)))

    image_row = 0
    image_col = 0
    for image in img_path:
        worksheet.insert_image(image_row,
                              image_col,
                              image,
                              {'x_scale': 0.5, 'y_scale': 0.5,
                              'x_offset': 5, 'y_offset': 5,
                              'positoning': 1})

        image_row += 1

#Product Title
    row = 0
    p_col = 1
    worksheet.set_column('B1:B5', 35)
    worksheet.set_default_row(250)
    for i,r in enumerate(product_title):
        p = r.find_all('h2')
        i = 0
        for c in p:
            pr_title = '{:20}'.format(c.text)
            prod_title.append(pr_title)
            listToStr = ' '.join(map(str, prod_title))
        pro_title.append(listToStr)
                #print()
        i+=1
        prod_title.clear()
        listToStr = ' '
        for z in range(0, len(pro_title)):
            worksheet.write(row, p_col, pro_title[z], cell_format)
            row+=1

#Product Price
    row = 0
    pr_col = 2
    worksheet.set_column('C1:C5', 10)
    worksheet.set_default_row(250)
    for i,r in enumerate(price):
        p = r.find_all('h2')
        i = 0
        for c in p:
            pr_price = '{:20}'.format(c.text)
            price_list.append(pr_price)
            listToStr = ' '.join(map(str, price_list))
        pr_list.append(listToStr)
                #print()
        i+=1
        price_list.clear()
        listToStr = ' '
        for z in range(0, len(pro_title)):
            worksheet.write(row, pr_col, pr_list[z], cell_format)
            row+=1

#Desc    
    row = 0
    d_col = 3
    worksheet.set_column('D1:D5', 100)
    worksheet.set_default_row(350)
    for i,r in enumerate(desc):
        p = r.find_all('div')
        i = 0
        for c in p:
            desc_data = '{:20}'.format(c.text)
            desc_list.append(desc_data)
            listToStr = ' '.join(map(str, desc_list))
        desc_li.append(listToStr)
                #print()
        i+=1
        desc_list.clear()
        listToStr = ' '
        for z in range(0, len(desc_li)):
            worksheet.write(row, d_col, desc_li[z], cell_format)
            row+=1
            
#Specs    
    row = 0
    e_col = 4
    worksheet.set_column('E1:E5', 100)
    worksheet.set_default_row(350)
    for i,r in enumerate(specs):
        p = r.find_all('table')
        i = 0
        for c in p:
            specs_data = '{:20}'.format(c.text)
            specs_list.append(specs_data)
            listToStr = ' '.join(map(str, specs_list))
        specs_li.append(listToStr)
                #print()
        i+=1
        specs_list.clear()
        listToStr = ' '
        for z in range(0, len(specs_li)):
            worksheet.write(row, e_col, specs_li[z], cell_format)
            row+=1

#Product Link
    row = 0
    link_col = 5
    worksheet.set_column('F1:F5', 40)
    worksheet.set_default_row(350)
    for z in range(30, 60):
        worksheet.write(row, link_col, url[z], cell_format)
        row+=1

workbook.close()
```
